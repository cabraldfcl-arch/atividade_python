# 1. Peça 5 números com for e diga quantos são pares

```python
pares = 0

for i in range(5):
    num = int(input("Digite um número: "))
    if num % 2 == 0:
        pares += 1

print("Quantidade de números pares:", pares)
```

# 2. Use while para pedir uma nota entre 0 e 10

```python
nota = -1

while nota < 0 or nota > 10:
    nota = float(input("Digite uma nota entre 0 e 10: "))

print("Nota válida:", nota)
```

# 3. Peça um número e mostre a tabuada com for

```python
num = int(input("Digite um número para ver a tabuada: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

# 4. Peça idade e documento. Use if com and para verificar acesso

```python
idade = int(input("Digite sua idade: "))
doc = input("Possui documento? (sim/não): ").lower()

if idade >= 18 and doc == "sim":
    print("Acesso liberado")
else:
    print("Acesso negado")
```

# 6. Sistema de login com 3 tentativas usando while

```python
usuario_correto = "admin"
senha_correta = "1234"

tentativas = 0

while tentativas < 3:
    usuario = input("Usuário: ")
    senha = input("Senha: ")

    if usuario == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso!")
        break
    else:
        tentativas += 1
        print("Dados incorretos. Tentativas restantes:", 3 - tentativas)

if tentativas == 3:
    print("Acesso bloqueado")
```
