Atividade I

```python
idade = int(input("Qual a sua idade: "))

if idade >= 12:
    ingresso = input("Tem ingresso (S/N): ").upper()
    if ingresso == "S":
        print("Entrada liberada")
    elif ingresso == "N":
        print("Compre ingresso")
    else:
        print("Resposta inválida")
else:
    print("Idade insuficiente")
```

Atividade II

```python
usuario = input("Digite o usuário: ")

if usuario == "admin":
    senha = input("Digite a senha: ")
    
    if senha == "1234":
        print("Acesso permitido")
    else:
        print("Senha incorreta")
else:
    print("Usuário inválido")
```

Atividade III

```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    frequencia = float(input("Digite a frequência (%): "))
    
    if frequencia >= 75:
        print("Aprovado")
    else:
        print("Reprovado por falta")
else:
    print("Reprovado por nota")
```

Atividade IV

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        print("Par positivo")
    else:
        print("Ímpar positivo")
else:
    print("Número não positivo")
```

Atividade V

```python
idade = int(input("Digite sua idade: "))

if idade >= 16:
    doc = input("Possui documento? (s/n): ").lower()
    
    if doc == "s":
        print("Pode entrar")
    else:
        print("Precisa apresentar documento")
else:
    print("Entrada não permitida")
```

Atividade VI

```python
valor = float(input("Digite o valor da compra: "))

if valor >= 100:
    cupom = input("Possui cupom? (s/n): ").lower()
    
    if cupom == "s":
        print("Desconto aplicado")
    else:
        print("Sem desconto")
else:
    print("Valor insuficiente para desconto")
```

Atividade VII

```python
numero = int(input("Digite um número inteiro: "))

if numero > 0:
    if numero % 2 == 0:
        if numero % 3 == 0:
            print("Múltiplo de 2 e de 3")
        else:
            print("Múltiplo de 2, mas não de 3")
    else:
        print("Não é múltiplo de 2")
else:
    print("Número não positivo")
```

Atividade VIII

```python
nome = input("Digite o nome: ").strip()

if nome != "":
    idade = int(input("Digite a idade: "))
    
    if idade >= 18:
        print("Cadastro completo")
    else:
        print("Cadastro incompleto: menor de idade")
else:
    print("Nome obrigatório")
```

Atividade IX

```python
idade = int(input("Digite a idade: "))

if idade >= 60:
    dor = input("Sente dor forte? (s/n): ").lower()
    
    if dor == "s":
        print("Atendimento prioritário imediato")
    else:
        print("Aguardar triagem")
else:
    print("Atendimento comum")
```
