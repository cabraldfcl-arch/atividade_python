1. Verificar número positivo

```python
numero = int(input("Digite um numero: "))

if numero > 0 :
    print("Numero positivo")
else:
    print("Numero negativo")
        
```

2. Verificar maior de idade

```python
idade = int(input("Digite sua idade: "))

if idade >= 18 :
    print("Maior de idade")
else:
    print("Menor de idade")
```

4. Par ou Ímpar

```python
numero = int(input("Digite um numero: "))

if numero % 2 == 0:
    print("Numero par")
else:
    print("numero impar")
```

6. Comparar dois números

```python
numero1 = int(input("Digite 1° Numero: "))
numero2 = int(input("Digite 2° Numero: "))

if numero1 > numero2:
    print("Numero 1° é Maior")
elif numero2 > numero1:
    print("Numero 2° é Maior")
```

9. Comparar três números

```python
n1 = int(input("Digite o primeiro número: "))
n2 = int(input("Digite o segundo número: "))
n3 = int(input("Digite o terceiro número: "))

maior = n1

if n2 > maior:
    maior = n2

if n3 > maior:
    maior = n3

print(f"O maior número é {maior}")
```

10. Sistema simples de login

```python
usuario_correto = "admin"
senha_correta = "1234"

usuario = input("Digite o usuário: ")
senha = input("Digite a senha: ")

if usuario == usuario_correto and senha == senha_correta:
    print("Login realizado com sucesso")
else:
    print("Usuário ou senha incorretos")

```
