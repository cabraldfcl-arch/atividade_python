2. Peça um número. Verifique se ele está entre 50 e 100.

```python
numero = float(input("Digite um número: "))

if numero >= 50:
    if numero <= 100:
        print("O número está entre 50 e 100")
    else:
        print("O número é maior que 100")
else:
    print("O número é menor que 50")
```

3. Peça o dia da semana. Mostre se é final de semana.

```python
dia = input("Digite o dia da semana: ")

if dia == "sábado":
    print("É final de semana")
else:
    if dia == "domingo":
        print("É final de semana")
    else:
        print("É dia útil")
```

4. Peça usuário e senha. Verifique se ambos estão corretos.

```python
usuario = input("Digite o usuário: ")

if usuario == "admin":
    senha = input("Digite a senha: ")
    if senha == "1234":
        print("Acesso liberado!")
    else:
        print("Senha incorreta!")
else:
    print("Usuário inexistente!")
```

5. Peça uma nota e a frequência do aluno. Mostre se ele foi aprovado.

```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7.0:
    frequencia = float(input("Digite a frequência do aluno (0 a 100): "))
    if frequencia >= 75:
        print("Aprovado!")
    else:
        print("Reprovado por frequência!")
else:
    print("Reprovado por nota!")
```

6. Peça se o aluno entregou o trabalho. Use not para verificar se ele não entregou.

```python
entregou = input("O aluno entregou o trabalho? (s/n): ")
if not entregou == "s":
    print("O aluno não entregou o trabalho!")
else:
    print("Trabalho entregue com sucesso.")
```

3. Peça números até o usuário digitar 0.

```python
numero = int(input("Digite um número (ou 0 para parar): "))

while numero != 0:
    print(f"Você digitou o número: {numero}")
    numero = int(input("Digite o próximo número (ou 0 para sair): "))

print("Programa encerrado.")
```

4. Conte de 10 até 1

```python
contador = 10

while contador >= 1:
    print(contador)
    contador = contador - 1

print("Fogo! 🚀")
```
